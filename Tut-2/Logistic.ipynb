```python
import numpy as np
import matplotlib.pyplot as plt
from numba import jit

plt.rcParams["figure.figsize"] = [12, 7]
```

# Linear Map

```python
n = 50
x1 = 0.337
x2 = 1e5
r1 = 1.3
r2 = 0.3
data1 = np.zeros((n))
data2 = np.zeros((n))
for i in range(n):
    data1[i] = x1
    data2[i] = x2
    x1 = r1 * x1
    x2 = r2 * x2
plt.plot(data1, "r", label="r = {}".format(r1))
plt.plot(data2, "b", label="r = {}".format(r2))
plt.legend(loc="best")
plt.title("LINEAR MAP", fontsize=15)
plt.xlabel(r"$n$", fontsize=15)
plt.ylabel(r"$x_{n}$", fontsize=15)
plt.show()
```

# Logistic Map

```python
n = 50
x = 0.337
r = [0.5, 2.5, 3.25, 3.5, 4]
data = np.zeros((n))
for j in range(len(r)):
    for i in range(n):
        data[i] = x
        x = r[j] * x * (1 - x)

    plt.plot(
        data,
        "r",
        marker="o",
        mfc="black",
        mec="none",
        label="r = {}".format(r[j]),
        ms=10,
    )
    plt.title("LOGISTIC MAP", fontsize=15)
    plt.legend(loc="best")
    plt.xlabel(r"$n$", fontsize=15)
    plt.ylabel(r"$x_{n}$", fontsize=15)
    data = np.zeros((n))
    x = 0.337
    plt.show()
```

```python
n = 100000
x = np.linspace(0, 1, n)
r = [0.5, 1, 2, 3, 4]
color = ["magenta", "green", "blue", "red", "yellow"]
plt.plot(x, x, "--", color="black", label="$x_{n+1}=x_{n}$")
for j in range(len(r)):
    data = r[j] * x * (1 - x)
    plt.plot(x, data, color[j], label="r = {}".format(r[j]))
    plt.title("LOGISTIC MAP", fontsize=15)
    plt.legend(loc="best")
    plt.xlabel(r"$x_{n}$", fontsize=15)
    plt.ylabel(r"$x_{n+1}$", fontsize=15)
plt.show()
```

```python
n = 100000
x = np.linspace(0, 1, n)
x_copy = np.copy(x)
r = [2, 3.2]
color = ["green", "magenta"]
data = np.zeros(np.size(x))
for j in range(len(r)):
    plt.plot(x_copy, x_copy, "--", color="black", label=r"$x_{n+2}=x_{n+1}=x_{n}$")
    for i in range(2):
        data = r[j] * x * (1 - x)
        x = r[j] * x * (1 - x)
        plt.plot(x_copy, data, color[i], label=r"$F^{}$ for r = {}".format(i + 1, r[j]))
        plt.legend(loc="best")
    plt.title("LOGISTIC MAP", fontsize=15)
    plt.xlabel(r"$x_{n}$", fontsize=15)
    plt.show()
    data = np.zeros((np.size(x)))
    x = np.linspace(0, 1, n)
```

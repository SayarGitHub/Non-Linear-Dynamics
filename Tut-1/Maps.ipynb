# Tent Map


```python
import numpy as np
import matplotlib.pyplot as plt
from numba import jit

plt.rcParams["figure.figsize"] = [12, 7]
```

```python
n = 100000
x = np.linspace(0, 1, n)
y = np.zeros((np.size(x)))
for i in range(len(x)):
    if x[i] < 0.5:
        y[i] = 2 * x[i]
    else:
        y[i] = 2 * (1 - x[i])
plt.plot(x, y, "r")
plt.title("TENT MAP", fontsize=15)
plt.xlabel(r"$x_n$", fontsize=15)
plt.ylabel(r"$x_{n+1}$", fontsize=15)
plt.show()
```

```python
y = np.linspace(0, 1, n)
for i in range(1, 5):
    x = np.linspace(0, 1, n)
    for j in range(i):
        for k in range(len(x)):
            if x[k] < 0.5:
                x[k] = 2 * x[k]
            else:
                x[k] = 2 * (1 - x[k])
    plt.plot(y, x, "r")
    plt.plot(y, y, "b")
    plt.title("TENT MAP for period " + str(i), fontsize=15)
    plt.xlabel(r"$x_n$", fontsize=15)
    plt.ylabel(f"$x_{{n+{i}}}$", fontsize=15)
    plt.show()
```

```python
n = 100
x = 0.013
y = np.zeros((n))
for i in range(n):
    if x < 0.5:
        y[i] = x
        x = 2 * x
    else:
        y[i] = x
        x = 2 * (1 - x)
plt.plot(y, "r", marker="o", mfc="black", mec="none", ms=10)
plt.title("TENT MAP Time Series", fontsize=15)
plt.xlabel(r"$n$", fontsize=15)
plt.ylabel(r"$x_{n+1}$", fontsize=15)
```

# Shift Map


```python
n = 100000
x = np.linspace(0, 1, n)
y = np.zeros(np.size(x))
for i in range(len(x)):
    y[i] = (2 * x[i]) % 1
plt.scatter(x, y, s=0.1)
plt.title("SHIFT MAP", fontsize=15)
plt.xlabel(r"$x_n$", fontsize=15)
plt.ylabel(r"$x_{n+1}$", fontsize=15)
plt.show()
```

```python
y = np.linspace(0, 1, n)
for i in range(1, 5):
    x = np.linspace(0, 1, n)
    for j in range(i):
        for k in range(len(x)):
            x[k] = (2 * x[k]) % 1

    plt.scatter(y, x, 0.1, "red")
    plt.plot(y, y, "b")
    plt.title("SHIFT MAP for period " + str(i), fontsize=15)
    plt.xlabel("$x_n$", fontsize=15)
    plt.ylabel(f"$x_{{n+{i}}}$", fontsize=15)
    plt.show()
```

```python
n = 100
x = 0.137
y = np.zeros((n))
for i in range(n):
    y[i] = x
    x = (2 * x) % 1
plt.plot(y, "r", marker="o", mfc="black", mec="none", ms=10)
plt.title("SHIFT MAP Time Series", fontsize=15)
plt.xlabel(r"$n$", fontsize=15)
plt.ylabel(r"$x_{n+1}$", fontsize=15)
```
